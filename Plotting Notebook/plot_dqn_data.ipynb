```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
episodes = []
episodes_200 = []
for i in range(1,1001):
    episodes.append(i)

for i in range(1,201):
    episodes_200.append(i)    
```

```python
random = np.loadtxt("random.txt")
distance = np.loadtxt("obs_distance.txt")
pos = np.loadtxt("obs_position.txt")
rew1 = np.loadtxt("reward_just_1.txt")
fiveh = np.loadtxt("winning_score_500.txt")
eps = np.loadtxt("epsilon_decay.txt")
learn = np.loadtxt("learning_rate.txt")
```

### random_play

```python
plt.plot(episodes, random, label="Zufälliges Spielen", color = '#ffa700', linewidth=2)
plt.xlabel('Episoden')
plt.ylabel('Score')
plt.title("Zufälliges Spielen")
plt.legend()
plt.savefig("random_play.png", figsize=(640,480), dpi=110)
plt.show()
```

### trained play

```python
plt.plot(episodes, random, label="Zufälliges Spielen", color = '#ffa700', linewidth=2)
plt.plot(episodes, distance, label="Spielen nach Training", color = '#2c89f2', linewidth=2)
plt.xlabel('Episoden')
plt.ylabel('Score')
plt.title("trainierter Agent")
plt.legend()
plt.savefig("trained_agent.png", figsize=(640,480), dpi=110)
plt.show()
```

### position observation

```python
plt.plot(episodes, distance, label="Spielen nach Training, Distanz Observation", color = '#2c89f2', linewidth=2)
plt.plot(episodes, pos, label="Spielen nach Training, Position Observation", color = '#58c900', linewidth=2)
plt.xlabel('Episoden')
plt.ylabel('Score')
plt.title("Vergleich Observationen")
plt.legend()
plt.savefig("compare_obs.png", figsize=(640,480), dpi=110)
plt.show()
```

### compare rewards

```python
plt.plot(episodes, pos, label="Spielen nach Training, Rewards für Gewinn/Niederlage", color = '#58c900', linewidth=2)
plt.plot(episodes, rew1, label="Spielen nach Training, Rewards nur für Überleben", color = '#810a6a', linewidth=2)
plt.xlabel('Episoden')
plt.ylabel('Score')
plt.title("Vergleich Rewards")
plt.legend()
plt.savefig("compare_rewards.png", figsize=(640,480), dpi=110)
plt.show()
```

### lower learning rate, bigger batch size

```python
plt.plot(episodes_200, learn, label="Spielen nach Training, learning Rate 0,01", color = '#810a6a', linewidth=2)
plt.xlabel('Episoden')
plt.ylabel('Score')
plt.title("Vergleich Learning Rate")
plt.legend()
plt.savefig("compare_lr.png", figsize=(640,480), dpi=110)
plt.show()
```
